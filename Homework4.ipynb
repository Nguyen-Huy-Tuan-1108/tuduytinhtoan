W4A20:Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n = int(input())
temp = n
if n <= 0:
    print(n, "không phải dạng 2^k")
else:
    while n % 2 == 0 and n > 1:
        n = n // 2
    if n == 1:
        print(temp, "là dạng 2^k")
    else:
        print(temp, "không phải dạng 2^k")
```

W4A21:Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input())
tong = 0
n = abs(n)
while n > 0:
    tong = tong + (n % 10)
    n = n // 10
print("Tổng là:", tong)
```

W4A22:Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = abs(int(input()))
chan = 0
le = 0
while n > 0:
    chu_so = n % 10
    if chu_so % 2 == 0:
        chan = chan + 1
    else:
        le = le + 1
    n = n // 10
print("Chẵn:", chan, "- Lẻ:", le)
```

W4A24:Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = float(input())
s = 0
n = 0
while s <= A:
    n = n + 1
    s = s + (1 / n)
print("Kết quả n nhỏ nhất là:", n)
```

W4A25:Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
so = int(input())
if so != -1:
    lon_nhat = so
    nho_nhat = so
    while True:
        so = int(input("Nhập tiếp (-1 để dừng): "))
        if so == -1:
            break
        if so > lon_nhat:
            lon_nhat = so
        if nho_nhat > so:
            nho_nhat = so
    print("Lớn nhất:", lon_nhat, "Nhỏ nhất:", nho_nhat)
```

W4A26:Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 


```python
A = int(input())
f1 = 1
f2 = 1
while f1 + f2 <= A:
    f_tiep_theo = f1 + f2
    f1 = f2
    f2 = f_tiep_theo
print(f2)
```

W4A27:Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng) Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
s = input()
so_tu = 0
dang_trong_tu = False
for char in s:
    if char != " ":
        if dang_trong_tu == False:
            so_tu = so_tu + 1
            dang_trong_tu = True
    else:
        dang_trong_tu = False
print(so_tu)
```

W4A29:Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành

```python
s = input()
ket_qua = ""
for char in s:
    if char.isupper():
        ket_qua = ket_qua + char.lower()
    else:
        ket_qua = ket_qua + char.upper()
print(ket_qua)
```

W4A30:Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính tổng 3 số nguyên đó 

```python
s = input()
tong = 0
so_tam = ""
for i in range(len(s)):
    char = s[i]
    if char != "," and char != " ":
        so_tam = so_tam + char
    if char == "," or i == len(s) - 1:
        if so_tam != "":
            tong = tong + int(so_tam)
            so_tam = ""
print(tong)
```

W4A31:Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số.

```python
s = input()
hoa = 0
thuong = 0
so = 0
for char in s:
    if char.isupper():
        hoa = hoa + 1
    elif char.islower():
        thuong = thuong + 1
    elif char.isdigit():
        so = so + 1
print("Hoa:", hoa, "- Thường:", thuong, "- Số:", so)
```
