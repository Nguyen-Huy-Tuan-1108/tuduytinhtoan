Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
a = float(input())
b = float(input())
c = float(input())

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        loai = "Tam giác Đều"
    elif a == b or b == c or a == c:
        loai = "Tam giác Cân"
    elif (a**2 + b**2 == c**2) or (a**2 + c**2 == b**2) or (b**2 + c**2 == a**2):
        loai = "Tam giác Vuông"
    else:
        loai = "Tam giác Thường"
    chu_vi = a + b + c
    p = chu_vi / 2
    dien_tich = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print(f"Đây là {loai}. Chu vi: {chu_vi}, Diện tích: {dien_tich:.2f}")
else:
    print("Khong la ba canh cua mot tam giac")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a = float(input())
b = float(input())
c = float(input())

if a == 0:
    if b == 0:
        print("Vô nghiệm" if c != 0 else "Vô số nghiệm")
    else:
        print("x =", -c / b)
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        print("Nghiệm kép x =", -b / (2 * a))
    else:
        x1 = (-b + delta**0.5) / (2 * a)
        x2 = (-b - delta**0.5) / (2 * a)
        print(f"Có 2 nghiệm: x1 = {x1}, x2 = {x2}")
```

Bài 8: Giải hệ phương trình tuyến tính:

```python
a, b, m, c, d, n = map(float, input().split())
D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c
if D != 0:
    print(f"Hệ có nghiệm duy nhất: x = {Dx/D}, y = {Dy/D}")
else:
    if Dx == 0 and Dy == 0:
        print("Hệ vô số nghiệm")
    else:
        print("Hệ vô nghiệm")
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây. 

```python
tong_giay = int(input())

gio = tong_giay // 3600
phut = (tong_giay % 3600) // 60
giay = tong_giay % 60

print(f"{gio} giờ, {phut} phút, {giay} giây")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào mộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
xtam = float(input())
ytam = float(input())
r = float(input())

xa = float(input())
ya = float(input())

khoangcach = ((xa - xtam)**2 + (ya - ytam)**2)**0.5

if khoangcach == r:
    print("Điểm A nằm trên đường tròn")
elif khoangcach < r:
    print("Điểm A nằm trong đường tròn")
else:
    print("Điểm A nằm ngoài đường tròn")
```
