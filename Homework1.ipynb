Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân. 

```python
a = float(input())
b = float(input())
dientichhcn = a * b
r = a / 2
dientichtron = 3.14 * (r ** 2)
dientichconlai = dientichhcn - dientichtron
print(f"{dientichconlai:.2f}")
```

Bài 4: Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c}
không phải là kí tự alphabet. 

```python
c = input()
if c.isalpha():
    print(f"'{c}' là ký tự alphabet")
else:
    print(f"'{c}' không phải là ký tự alphabet")
```

Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 

```python
chu_hoa = input("Nhập vào chữ cái hoa: ")
chu_thuong = chu_hoa.lower()
if chu_thuong == 'a':
    ket_qua = 'z'
else:
    ket_qua = chr(ord(chu_thuong) - 1)
print("Chữ cái thường liền trước là:", ket_qua)
```

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
a = float(input())
b = float(input())
c = float(input())
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    s = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print(f"Diện tích tam giác là: {s:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```
