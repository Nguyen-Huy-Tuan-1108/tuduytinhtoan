W4A6:Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b)."

```python
a=int(input())
b=int(input())
tong=0
for n in range(a,b+1):
    if n<2:
        continue
    for i in range(2,n):
        if n%i==0:
            break
else: tong=tong+n
print(tong)
```

W4A7:Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 

```python
n=int(input())
uocmax=0
for i in range(2,n+1):
    if n%i==0: 
        for j in range (2,i):
            if i%j==0:
                break
        else: uocmax=i
print(uocmax)
    
```

W4A8:Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n=int(input())
dem=0
while 1:
    rev=int(str(n)[::-1])
    if n==rev:
        break
    n=n+rev
    dem=dem+1
print(dem,n)
```

W4A9:Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number).

```python
n = int(input("Nhập n: "))
ket_qua = []
i = 1
while i * i <= n:
    so_cp = i * i
    s_cp = str(so_cp) 
    if len(s_cp) == len(set(s_cp)):
        ket_qua.append(s_cp)
    i += 1
if len(ket_qua) == 0:
    print("(no number)")
else:
    print(" ".join(ket_qua))
```

W4A10:Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số

W4A13:Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
a=int(input())
b=int(input())
tonga=0
tongb=0
for i in range(1,a):
    if a%i==0:
        tonga=tonga+i
for i in range (1,b):
    if b%i==0:
        tongb=tongb+i
if tonga==tongb:
    print("True")
else: print("False")
```
