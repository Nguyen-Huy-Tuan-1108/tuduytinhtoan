W5A1:Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> int


```python
def Max_Of_Two(a, b):
    if a > b:
        return a
    else:
        return b
print(Max_Of_Two(3,7))
```

W5A3:Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 

```python
def nguyento(n):
    if n<2:
        return False   
    for i in range(2,n):
        if n%i==0:
            return False
        else:
            return True
print(nguyento(7))
```

W5A4:VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def sohoanhao(n):
    if n<=1:
        return False
    tong=0
    for i in range(1,n):
        if n%i==0:
            tong=tong+i
    return tong==n
print(sohoanhao(10))
```

W5A5:Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử 

```python
def tim(list,k):
    for i in range(len(list)):
        if list[i]==k:
            return i
    return -1
print(tim([1,2,3,4],3))
```

W5A6:Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)

```python
def giaithua(n):
    tich=1
    for i in range(1,n+1):
        tich=tich*i
    return tich
n=int(input())
print(giaithua(n))
```

W5A8:
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.

```python
def hamming_distance(a, b):
    x = a ^ b
    count = 0

    while x > 0:
        if x % 2 == 1:
            count += 1
        x //= 2
    return count
a=int(input())
b=int(input())
print(hamming_distance(a,b))
```

W5A12:Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 

```python
def divide_with_remainder(dividend, divisor):
    q = dividend // divisor
    r = dividend % divisor
    return q, r
print(divide_with_remainder(10, 3))
```

W5A14:Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op.

W5A15:Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a, b, c):
    if a*a + b*b == c*c:
        return True
    elif a*a + c*c == b*b:
        return True
    elif b*b + c*c == a*a:
        return True
    else:
        return False
print(is_right_triangle(3, 4, 5)) 
```
